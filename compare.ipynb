```python
import os
import matplotlib.pyplot as plt

# Kiểm tra và tạo thư mục './time' nếu chưa tồn tại
if not os.path.exists('./time'):
    os.makedirs('./time')

# Dữ liệu
thresholds = [100, 300, 500, 700, 900]

# Foodmart
tuhufp_fm_times = [2, 2, 4, 7, 8]
tuhufp_growth_fm_times = [36, 102, 185, 311, 431]

# Chess
tuhufp_chess_times = [68, 318, 523, 698, 919]
tuhufp_growth_chess_times = [332, 1541, 2665, 3758, 5238]

# Retail
tuhufp_retail_times = [467, 590, 799, 997, 1214]
tuhufp_growth_retail_times = [280, 305, 449, 466, 527]

# Vẽ và lưu biểu đồ Foodmart vào thư mục './time' và hiển thị
plt.figure(figsize=(10, 6))
plt.plot(thresholds, tuhufp_fm_times, marker='o', label='TUHUFP - Foodmart')
plt.plot(thresholds, tuhufp_growth_fm_times, marker='s', label='TUHUFP Growth - Foodmart')
plt.title('Comparison of TUHUFP and TUHUFP Growth (Foodmart)', fontsize=14)
plt.xlabel('Threshold', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('./time/foodmart_comparison.png')  # Lưu biểu đồ vào thư mục './time'
plt.show()  # Hiển thị biểu đồ trên màn hình
plt.close()

# Vẽ và lưu biểu đồ Chess vào thư mục './time' và hiển thị
plt.figure(figsize=(10, 6))
plt.plot(thresholds, tuhufp_chess_times, marker='o', label='TUHUFP - Chess')
plt.plot(thresholds, tuhufp_growth_chess_times, marker='s', label='TUHUFP Growth - Chess')
plt.title('Comparison of TUHUFP and TUHUFP Growth (Chess)', fontsize=14)
plt.xlabel('Threshold', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('./time/chess_comparison.png')  # Lưu biểu đồ vào thư mục './time'
plt.show()  # Hiển thị biểu đồ trên màn hình
plt.close()

# Vẽ và lưu biểu đồ Retail vào thư mục './time' và hiển thị
plt.figure(figsize=(10, 6))
plt.plot(thresholds, tuhufp_retail_times, marker='o', label='TUHUFP - Retail')
if all(v is not None for v in tuhufp_growth_retail_times):  # Kiểm tra dữ liệu của TUHUFP Growth
    plt.plot(thresholds, tuhufp_growth_retail_times, marker='s', label='TUHUFP Growth - Retail')
else:
    print("No data available for TUHUFP Growth - Retail.")
plt.title('Comparison of TUHUFP and TUHUFP Growth (Retail)', fontsize=14)
plt.xlabel('Threshold', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('./time/retail_comparison.png')  # Lưu biểu đồ vào thư mục './time'
plt.show()  # Hiển thị biểu đồ trên màn hình
plt.close()
```
